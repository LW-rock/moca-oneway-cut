## 1. Import Local Package and Example Script

Run this notebook from either the package root or the repository root. The local `src` directory is added automatically when the package has not been installed yet.


```python
import sys
from pathlib import Path

cwd = Path.cwd()
package_root = cwd if (cwd / "src" / "moca_oneway_cut").exists() else cwd / "moca-oneway-cut"
if not (package_root / "src" / "moca_oneway_cut").exists():
    package_root = cwd.parent if (cwd.parent / "src" / "moca_oneway_cut").exists() else cwd

src_dir = package_root / "src"
examples_dir = package_root / "examples"
for path in (src_dir, examples_dir):
    if path.exists() and str(path) not in sys.path:
        sys.path.insert(0, str(path))

#print(f"Package root: {package_root}")
```
